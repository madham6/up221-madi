I am going to look at the variable of 'No Vehicle' to understand each household's level of car access (the distinction being between no cars and at least one car). The theory is that food access, beyond proximity to a location selling or distributing fresh food, is heavily dependent on the means of transportation. We know that low-income households are more likely to be zero vehicle households and live in food deserts so we may want to explore this relationship as part of our project.

Taking a quick look at the percentage of households in each census tract with zero vehicle access, we can see the median is around nine percent and the mean is around 11.5. Next, I will plot a histogram of this variable to better understand the total distribution.

```python
census2['Percent No Vehicle'].plot.hist(bins=50)
```

It looks like this is a positively skewed distribution where the most census tracts have a close to 0% proportion of households with zero access to vehicles. However, there are some outliers where over 50% of households do not own cars. If they lack a close fresh food option, their access to food is much lower. Perhaps we should explore the relationship between vehicle access and race to see if there are any trends.
