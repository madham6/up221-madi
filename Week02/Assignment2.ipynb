Lots of columns have null values, so this dataset is relatively incomplete. Other than that, the data types are mostly strings with a few integers and floats mixed in and geometry (which contains location data).

I want to look at the number of resources that were added by year, so first I must create a year variable. The date updated variable includes months and days which would make too many buckets when plotted as a bar chart.

```python
food['yr'] = [x[:4] for x in food['date_updat']]
food
```

It looks like the year variable works in only including the first four characters of the date_updat variable.
Now, I want to see a summary of how many resources were added per year before creating a bar chart. I also want to make this a series so I can plot based on the series.

```python
food['yr'].value_counts()
```

```python
yr_count = food['yr'].value_counts()
yr_count
```

```python
yr_count - yr_count.reset_index()
yr_count
```

```python
type(yr_count)
```

I just checked that this is indeed a series and I can create the bar chart now.

```python
yr_count.plot.bar(x = 'Year', y = 'count', title = 'New Resources by Year')
```

If I had more time, I would figure out a way to sort the x-axis according to the year rather than the frequency. For now, this is a pretty good visual representation of which years saw the most growth in resources. I also need to check with the metadata to understand if the date_updat variable refers to the date the resource was brought online or if that's the date someone added the row to the dataset.
Next, I'm going to keep only the necessary columns (the ones with no null values or with important information for later analysis).

```python
new_food = food[['OBJECTID', 'source', 'cat1', 'cat2', 'org_name', 'Name', 'addrln1', 'addrln2', 'city', 'state', 'hours', 'phones', 'url', 'post_id', 'descriptio', 'zip', 'link', 'use_type', 'latitude', 'longitude', 'date_updat', 'POINT_X', 'POINT_Y', 'geometry', 'yr']]
new_food
```

Here I am constraining the dataset to just the resources added in 2013 for my map.

```python
food13 = new_food[new_food.yr == '2013']
food13
```

```python
food13.geometry
```

Here I create my map using the open streets map basemap. Later on I will constrain the location to Los Angeles using the average longitude and latitude of the data points so the map is zoomed in more.

```python
m = folium.Map()
m
```

```python
latitude = food13.latitude.mean()
latitude
```

```python
longitude = food13.longitude.mean()
longitude
```

```python
m = folium.Map(location=[latitude,longitude])
m
```

Now that the basemap is zoomed in on the correct location I want to assign the location and name of each resource to be displayed when the cursor moves over the datapoint or clicks on it.

```python
for index, row in food13.iterrows():
    print(row.Name, row.latitude, row.longitude)
```

```python
for index, row in food13.iterrows():
    folium.Marker([row.latitude, row.longitude], popup=row.Name, tooltip=row.Name).add_to(m)
m
```

```python
m.save('food13.html')
```
